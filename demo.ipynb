# StoPharma: End-to-End Demo
Questo notebook mostra il flusso completo di esplorazione, dall'ingestion del dataset fino alla validazione farmacologica (Fasi 1-4).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hopfield_core import energy, glauber_step, overlap, update_sync
```

## 1. Caricamento Dati e Pesi
Carichiamo la rete di Hopfield pre-addestrata (Fase 2) con il theta ottimizzato per la sparsità target.

```python
patterns = np.load('data/patterns.npy').astype(np.float64)
W = np.load('data/W.npy')
theta = float(np.load('data/theta.npy'))
molecules = pd.read_csv('data/molecules.csv')
names = molecules['name'].tolist()
print(f"Rete caricata: N={W.shape[0]} nodi, P={patterns.shape[0]} pattern noti.\nTheta calibrato: {theta:.2f}")
```

## 2. Generazione Creativa (Simulated Annealing)
Iniettiamo rumore termico e forziamo la rete a rilassarsi in uno stato genuinamente nuovo (stato spurio).

```python
S = np.random.choice([-1.0, 1.0], len(names))
T = 2.0
T_end = 0.05
cooling = 0.995
history = []
while T > T_end:
    for _ in range(50):
        S = glauber_step(W, S, T, theta=theta)
    history.append({'T': T, 'E': energy(W, S, theta=theta)})
    T *= cooling
S_final = S
active_mols = [names[i] for i in range(len(names)) if S_final[i] == 1]
print(f"Cocktail generato ({len(active_mols)} principi attivi):\n{active_mols}")
```

## 3. Curva Termodinamica
Visualizziamo la discesa verso il minimo energetico.

```python
plt.figure(figsize=(9,5))
plt.plot([h['T'] for h in history], [h['E'] for h in history], marker='.', alpha=0.7)
plt.gca().invert_xaxis()
plt.title('Discesa dell\'Energia (Simulated Annealing)')
plt.xlabel('Temperatura (rumore decrescente)')
plt.ylabel('Energia E')
plt.grid(True)
plt.show()
```
